```python
my_list = list(df["appliances"].unique())
```

```python
my_list
```

```python
def cleaning_appliances(x):
    if x==my_list[0]:
        return 'Laundry (In Building)','Fridge / Freezer'
    elif x==my_list[1]:
        return 'Laundry (In Unit)', 'Dishwasher', 'Fridge / Freezer'
    elif x==my_list[2]:
        return 'Not Included'
    elif x==my_list[3]:
        return 'Laundry (In Building)'
    elif x==my_list[4]:
        return 'Laundry (In Unit)'
    elif x==my_list[5]:
        return 'Laundry (In Unit)','Fridge / Freezer'
    elif x==my_list[6]:
        return 'Laundry (In Unit)','Laundry (In Building)','Dishwasher','Fridge / Freezer'
    elif x==my_list[7]:
        return 'Laundry (In Building)', 'Dishwasher', 'Fridge / Freezer'
    elif x==my_list[8]:
        return 'Fridge / Freezer'
    elif x==my_list[9]:
        return 'Laundry (In Building)', 'Dishwasher'
    elif x==my_list[10]:
        return 'Dishwasher', 'Fridge / Freezer'
    elif x==my_list[11]:
        return 'Laundry (In Unit)', 'Laundry (In Building)', 'Fridge / Freezer'
    elif x==my_list[12]:
        return 'Laundry (In Unit)', 'Dishwasher'
    elif x==my_list[13]:
        return 'Dishwasher'
    elif x==my_list[14]:
        return 'Laundry (In Unit)', 'Laundry (In Building)', 'Dishwasher'
    elif x==my_list[15]:
        return 'Laundry (In Unit)', 'Laundry (In Building)'
```

```python
df["appliances2"] = df["appliances"].apply(lambda x: cleaning_appliances(x))
```

Now we are making colums for each params included yes or no

```python
def Dishwasher(x):
    try:
        x.index('Dishwasher')
        return "yes"
    except:
        return "no"
```

```python
def Laundry_In_Unit(x):
    try:
        x.index('Laundry (In Unit)')
        return "yes"
    except:
        return "no"
```

```python
def Fridge_Freezer(x):
    try:
        x.index('Fridge / Freezer')
        return "yes"
    except:
        return "no"
```

```python
def Laundry_In_Building(x):
    try:
        x.index('Laundry (In Building)')
        return "yes"
    except:
        return "no"
```

```python
df["Laundry_In_Building"] = df["appliances2"].apply(lambda x: Laundry_In_Building(x))
df["Fridge_Freezer"] = df["appliances2"].apply(lambda x: Fridge_Freezer(x))
df["Laundry_In_Unit"] = df["appliances2"].apply(lambda x: Laundry_In_Unit(x))
df["Dishwasher"] = df["appliances2"].apply(lambda x: Dishwasher(x))
```

```python
# as you can see we can change not include to no 
df['Hydro'] = df['Hydro'].map({'Not Included':'no',"Yes":"yes","No":"no"})
df['Water'] = df['Water'].map({'Not Included':'no',"Yes":"yes","No":"no"})
df['Heat'] = df['Heat'].map({'Not Included':'no',"Yes":"yes","No":"no"})
```

```python
# then we are creating def function for compine all this staff in one columns
def hydro_water_heat(df):
    if df["Hydro"]=="yes" and df["Water"]=="yes" and df["Heat"]=="yes":
        return "All Utilities Included"
    elif df["Water"]=="yes" and df["Heat"]=="yes" and df["Hydro"]=="no":
        return "Water and Heat Included"
    elif df["Hydro"]=="yes" and df["Water"]=="yes" and df["Heat"]=="no":
        return "Hydro and Water Included" 
    elif df["Hydro"]=="yes" and df["Heat"]=="yes" and df["Water"]=="no":
        return "Hydro and Heat Included"
    elif df["Hydro"]=="yes" and df["Water"]=="no" and df["Heat"]=="no":
        return "Hydro Included"
    elif df["Hydro"]=="no" and df["Water"]=="yes" and df["Heat"]=="no":
        return "Water Included"
    elif df["Hydro"]=="no" and df["Water"]=="no" and df["Heat"]=="yes":
        return "Heat Included"
    elif df["Hydro"]=="no" and df["Water"]=="no" and df["Heat"]=="no":
        return "Not Included"
```

```python
df["hydro_water_heat"] = df.apply(hydro_water_heat,axis=1)
```

```python
#Then checking result to preventing errors 
a = 3
print("result of test {} , hydro {} , water {} , heat {}".format(df["hydro_water_heat"][a],df["Hydro"][a],df["Water"][a],df["Heat"][a]))
```

```python
def price(x):
    try: #if price provided
        a = x.split(",") #exm we need split 1,400 2 part
        b = a[0] + a[1] # then compine it 
        return int(b) #return it b
    except:# if not privided
        return None
```
